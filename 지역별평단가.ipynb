```python
import tkinter as tk
from tkinter import ttk, filedialog, messagebox
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import matplotlib.font_manager as fm
from datetime import datetime
import numpy as np
import json
import os
import openpyxl
import warnings
import re  # ê¸°ì¡´ import êµ¬ë¬¸ë“¤ê³¼ í•¨ê»˜ ì¶”ê°€
import tkinter as tk
from tkinter import ttk, messagebox, filedialog
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from PIL import Image, ImageTk
import pandas as pd
import numpy as np
import os
import json
import requests
import xml.etree.ElementTree as ET
from datetime import datetime, timedelta
import logging
import shutil
from matplotlib import font_manager
import time
import concurrent.futures
import threading
# ì¶”ê°€í•  ì„í¬íŠ¸
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException, NoSuchElementException

# openpyxl ê²½ê³  ì–µì œ
warnings.filterwarnings('ignore', category=UserWarning, module='openpyxl')

class ApartmentPriceAnalyzer:
    def __init__(self, root):
        self.root = root
        self.root.title("ì•„íŒŒíŠ¸ ë§¤ë§¤ê°€ ìƒìŠ¹/í•˜ë½ë¥  ë¶„ì„")
        self.root.geometry("1200x800")
        
        # ì„¤ì • íŒŒì¼ ê²½ë¡œ
        self.config_file = "apartment_analyzer_config.json"
        
        # í•œê¸€ í°íŠ¸ ì„¤ì •
        plt.rcParams['font.family'] = ['Malgun Gothic', 'AppleGothic', 'Noto Sans CJK KR']
        plt.rcParams['axes.unicode_minus'] = False
        
        self.df = None
        self.create_widgets()
        
        # í”„ë¡œê·¸ë¨ ì¢…ë£Œ ì‹œ ì„¤ì • ì €ì¥
        self.root.protocol("WM_DELETE_WINDOW", self.on_closing)
        
        # ì‹œì‘ ì‹œ ì´ì „ ì„¤ì • ë¡œë“œ
        self.load_settings()
        
    def create_widgets(self):
        # ë©”ì¸ í”„ë ˆì„
        main_frame = ttk.Frame(self.root, padding="10")
        main_frame.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))
        
        # íŒŒì¼ ì„ íƒ í”„ë ˆì„
        file_frame = ttk.LabelFrame(main_frame, text="ì—‘ì…€ íŒŒì¼ ì„ íƒ", padding="10")
        file_frame.grid(row=0, column=0, columnspan=2, sticky=(tk.W, tk.E), pady=(0, 10))
        
        self.file_path_var = tk.StringVar()
        ttk.Entry(file_frame, textvariable=self.file_path_var, width=60, state="readonly").grid(row=0, column=0, padx=(0, 10))
        ttk.Button(file_frame, text="íŒŒì¼ ì„ íƒ", command=self.select_file).grid(row=0, column=1)
        ttk.Button(file_frame, text="ë°ì´í„° ë¡œë“œ", command=self.load_data).grid(row=0, column=2, padx=(10, 0))
        
        # ì§€ì—­ ì„ íƒ í”„ë ˆì„
        region_frame = ttk.LabelFrame(main_frame, text="ì§€ì—­ ì„ íƒ ë° ì°¨íŠ¸ ì˜µì…˜", padding="10")
        region_frame.grid(row=1, column=0, sticky=(tk.W, tk.E, tk.N, tk.S), pady=(0, 10))
        
        # ì§€ì—­ ì„ íƒ ë°©ì‹ ì„ íƒ
        ttk.Label(region_frame, text="ì„ íƒ ë°©ì‹:").grid(row=0, column=0, sticky=tk.W, pady=(0, 5))
        self.selection_mode = tk.StringVar(value="single")
        ttk.Radiobutton(region_frame, text="ë‹¨ì¼ ì„ íƒ", variable=self.selection_mode, value="single", 
                       command=self.on_selection_mode_change).grid(row=1, column=0, sticky=tk.W)
        ttk.Radiobutton(region_frame, text="ë‹¤ì¤‘ ì„ íƒ", variable=self.selection_mode, value="multiple",
                       command=self.on_selection_mode_change).grid(row=2, column=0, sticky=tk.W)
        
        # ì°¨íŠ¸ ìœ í˜• ì„ íƒ ì¶”ê°€
        ttk.Label(region_frame, text="ì°¨íŠ¸ ìœ í˜•:").grid(row=3, column=0, sticky=tk.W, pady=(10, 5))
        self.chart_type = tk.StringVar(value="change_rate")
        ttk.Radiobutton(region_frame, text="ìƒìŠ¹/í•˜ë½ë¥  (%)", variable=self.chart_type, value="change_rate").grid(row=4, column=0, sticky=tk.W)
        ttk.Radiobutton(region_frame, text="ì ˆëŒ€ê°€ê²© (ë§Œì›/í‰)", variable=self.chart_type, value="absolute_price").grid(row=5, column=0, sticky=tk.W)
        
        # ê¸°ì¤€ì  ì„ íƒ ì¶”ê°€
        ttk.Label(region_frame, text="ê¸°ì¤€ì  (ìƒìŠ¹/í•˜ë½ë¥  ê³„ì‚°ìš©):").grid(row=6, column=0, sticky=tk.W, pady=(10, 5))
        self.base_point = tk.StringVar(value="first")
        ttk.Radiobutton(region_frame, text="ì²« ë²ˆì§¸ ë°ì´í„°", variable=self.base_point, value="first").grid(row=7, column=0, sticky=tk.W)
        ttk.Radiobutton(region_frame, text="ì´ì „ ê¸°ê°„ ëŒ€ë¹„", variable=self.base_point, value="previous").grid(row=8, column=0, sticky=tk.W)
        ttk.Radiobutton(region_frame, text="íŠ¹ì • ì—°ë„ ê¸°ì¤€", variable=self.base_point, value="year").grid(row=9, column=0, sticky=tk.W)
        
        # ê¸°ì¤€ ì—°ë„ ì…ë ¥
        year_frame = ttk.Frame(region_frame)
        year_frame.grid(row=10, column=0, sticky=(tk.W, tk.E), pady=(5, 10))
        ttk.Label(year_frame, text="ê¸°ì¤€ ì—°ë„:").pack(side=tk.LEFT)
        self.base_year_var = tk.StringVar(value="2020")
        year_entry = ttk.Entry(year_frame, textvariable=self.base_year_var, width=8)
        year_entry.pack(side=tk.LEFT, padx=(5, 0))
        
        # ë‹¨ì¼ ì„ íƒìš© ì½¤ë³´ë°•ìŠ¤
        ttk.Label(region_frame, text="ì§€ì—­:").grid(row=11, column=0, sticky=tk.W, pady=(10, 5))
        self.region_var = tk.StringVar()
        self.region_combo = ttk.Combobox(region_frame, textvariable=self.region_var, state="readonly", width=30)
        self.region_combo.grid(row=12, column=0, sticky=(tk.W, tk.E), pady=(0, 10))
        
        # ë‹¤ì¤‘ ì„ íƒìš© ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤
        ttk.Label(region_frame, text="ì§€ì—­ ëª©ë¡ (Ctrl+í´ë¦­ìœ¼ë¡œ ë‹¤ì¤‘ ì„ íƒ):").grid(row=13, column=0, sticky=tk.W, pady=(10, 5))
        
        # ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ í”„ë ˆì„
        listbox_frame = ttk.Frame(region_frame)
        listbox_frame.grid(row=14, column=0, sticky=(tk.W, tk.E, tk.N, tk.S), pady=(0, 10))
        
        # ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ì™€ ìŠ¤í¬ë¡¤ë°”
        self.region_listbox = tk.Listbox(listbox_frame, selectmode=tk.EXTENDED, height=8, width=30)
        scrollbar = ttk.Scrollbar(listbox_frame, orient=tk.VERTICAL, command=self.region_listbox.yview)
        self.region_listbox.configure(yscrollcommand=scrollbar.set)
        
        self.region_listbox.grid(row=0, column=0, sticky=(tk.W, tk.E, tk.N, tk.S))
        scrollbar.grid(row=0, column=1, sticky=(tk.N, tk.S))
        
        listbox_frame.columnconfigure(0, weight=1)
        listbox_frame.rowconfigure(0, weight=1)
        
        # ì„ íƒëœ ì§€ì—­ í‘œì‹œ
        ttk.Label(region_frame, text="ì„ íƒëœ ì§€ì—­:").grid(row=15, column=0, sticky=tk.W, pady=(10, 5))
        self.selected_regions_var = tk.StringVar(value="ì—†ìŒ")
        selected_label = ttk.Label(region_frame, textvariable=self.selected_regions_var, wraplength=300)
        selected_label.grid(row=16, column=0, sticky=(tk.W, tk.E), pady=(0, 10))
        
        # ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ ì„ íƒ ì´ë²¤íŠ¸ ë°”ì¸ë”©
        self.region_listbox.bind('<<ListboxSelect>>', self.on_listbox_select)
        
        # ê·¸ë˜í”„ ê·¸ë¦¬ê¸° ë²„íŠ¼
        ttk.Button(region_frame, text="ê·¸ë˜í”„ ê·¸ë¦¬ê¸°", command=self.plot_graph).grid(row=17, column=0, pady=10)
        
        # ì´ˆê¸°ì—ëŠ” ë‹¨ì¼ ì„ íƒ ëª¨ë“œë¡œ ì„¤ì •
        self.on_selection_mode_change()
        
        # ê·¸ë˜í”„ í”„ë ˆì„
        self.graph_frame = ttk.LabelFrame(main_frame, text="ì‹œê³„ì—´ ê·¸ë˜í”„", padding="10")
        self.graph_frame.grid(row=1, column=1, sticky=(tk.W, tk.E, tk.N, tk.S), padx=(10, 0))
        
        # ê·¸ë¦¬ë“œ ê°€ì¤‘ì¹˜ ì„¤ì •
        main_frame.columnconfigure(1, weight=1)
        main_frame.rowconfigure(1, weight=1)
        self.root.columnconfigure(0, weight=1)
        self.root.rowconfigure(0, weight=1)
            
    def calculate_change_rate(self, data, base_point_type="first", base_year=None):
        """ìƒìŠ¹/í•˜ë½ë¥  ê³„ì‚°"""
        if len(data) == 0:
            return pd.Series(dtype=float), None
        
        base_date = None
        
        if base_point_type == "first":
            # ì²« ë²ˆì§¸ ë°ì´í„° ëŒ€ë¹„ ë³€í™”ìœ¨
            base_value = data.iloc[0]
            base_date = data.index[0]
            if base_value == 0 or pd.isna(base_value):
                return pd.Series([0] * len(data), index=data.index), base_date
            change_rate = ((data - base_value) / base_value * 100)
        elif base_point_type == "previous":
            # ì´ì „ ê¸°ê°„ ëŒ€ë¹„ ë³€í™”ìœ¨
            change_rate = data.pct_change() * 100
            change_rate.iloc[0] = 0  # ì²« ë²ˆì§¸ ê°’ì€ 0ìœ¼ë¡œ ì„¤ì •
            base_date = None
        else:  # year
            # íŠ¹ì • ì—°ë„ ê¸°ì¤€ ë³€í™”ìœ¨
            try:
                base_year_int = int(base_year) if base_year else 2020
                
                # í•´ë‹¹ ì—°ë„ì˜ ì²« ë²ˆì§¸ ë°ì´í„°ë¥¼ ì°¾ê¸°
                year_mask = data.index.year == base_year_int
                if not year_mask.any():
                    # í•´ë‹¹ ì—°ë„ê°€ ì—†ìœ¼ë©´ ê°€ì¥ ê°€ê¹Œìš´ ì—°ë„ ì°¾ê¸°
                    available_years = data.index.year.unique()
                    closest_year = min(available_years, key=lambda x: abs(x - base_year_int))
                    year_mask = data.index.year == closest_year
                    print(f"ê¸°ì¤€ ì—°ë„ {base_year_int}ê°€ ì—†ì–´ì„œ {closest_year}ë…„ì„ ì‚¬ìš©í•©ë‹ˆë‹¤.")
                
                base_idx = data[year_mask].index[0]
                base_value = data.loc[base_idx]
                base_date = base_idx
                
                if base_value == 0 or pd.isna(base_value):
                    return pd.Series([0] * len(data), index=data.index), base_date
                
                # ê¸°ì¤€ ì—°ë„ ì´í›„ ë°ì´í„°ë§Œ ê³„ì‚°
                data_after_base = data[data.index >= base_idx]
                change_rate = ((data_after_base - base_value) / base_value * 100)
                
            except (ValueError, IndexError):
                # ì˜¤ë¥˜ ë°œìƒ ì‹œ ì²« ë²ˆì§¸ ë°ì´í„° ê¸°ì¤€ìœ¼ë¡œ fallback
                base_value = data.iloc[0]
                base_date = data.index[0]
                if base_value == 0 or pd.isna(base_value):
                    return pd.Series([0] * len(data), index=data.index), base_date
                change_rate = ((data - base_value) / base_value * 100)
        
        return change_rate, base_date
    
    def read_monthly_data_with_openpyxl(self, file_path, sheet_name):
        """openpyxlì„ ì‚¬ìš©í•´ì„œ ì›”ë³„ ë°ì´í„° ì§ì ‘ ì¶”ì¶œ - ê°œì„ ëœ ë²„ì „"""
        import openpyxl
        
        print("=" * 60)
        print("ì›”ë³„ ë°ì´í„° ì§ì ‘ ì¶”ì¶œ ì‹œì‘ (ê°œì„ ëœ ë²„ì „)")
        print("=" * 60)
        
        # ì—‘ì…€ íŒŒì¼ ì—´ê¸°
        workbook = openpyxl.load_workbook(file_path, data_only=True)
        worksheet = workbook[sheet_name]
        
        print(f"ì›Œí¬ì‹œíŠ¸ ì •ë³´: {worksheet.max_row}í–‰ x {worksheet.max_column}ì—´")
        
        # 2~4í–‰ì—ì„œ ì§€ì—­ëª… ì¶”ì¶œ
        regions = []
        
        print("\nì§€ì—­ëª… ì¶”ì¶œ ì¤‘...")
        for col in range(2, min(worksheet.max_column + 1, 50)):  # Bì—´ë¶€í„° ìµœëŒ€ 50ê°œ ì»¬ëŸ¼
            # 2í–‰: ì‹œë„
            sido_cell = worksheet.cell(row=2, column=col)
            sido = str(sido_cell.value).strip() if sido_cell.value else ""
            
            # 3í–‰: ì‹œêµ°êµ¬  
            sigungu_cell = worksheet.cell(row=3, column=col)
            sigungu = str(sigungu_cell.value).strip() if sigungu_cell.value else ""
            
            # 4í–‰: ì¶”ê°€ ì •ë³´ (ìˆìœ¼ë©´)
            extra_cell = worksheet.cell(row=4, column=col)
            extra = str(extra_cell.value).strip() if extra_cell.value else ""
            
            # ì§€ì—­ëª… ì¡°í•©
            if sido and sigungu:
                if extra and extra not in sido and extra not in sigungu and len(extra) < 10:
                    region_name = f"{sido} {sigungu} {extra}".strip()
                else:
                    region_name = f"{sido} {sigungu}".strip()
                
                # ìœ íš¨í•œ ì§€ì—­ëª…ì¸ì§€ í™•ì¸
                exclude_terms = ['êµ¬ë¶„', 'ì§€ì—­', 'ì‹œì ', 'ë…„ë„', 'í‰ê· ', 'ê³„', 'total']
                if not any(term in region_name.lower() for term in exclude_terms):
                    regions.append((col, region_name))
                    
                    if len(regions) <= 10:  # ì²˜ìŒ 10ê°œë§Œ ì¶œë ¥
                        print(f"  ì»¬ëŸ¼ {col}: {region_name}")
        
        print(f"ì´ {len(regions)}ê°œ ì§€ì—­ ë°œê²¬")
        
        if len(regions) == 0:
            raise ValueError("ìœ íš¨í•œ ì§€ì—­ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        
        # 5í–‰ë¶€í„° ë°ì´í„° ëê¹Œì§€ ëª¨ë“  í–‰ì„ í™•ì¸
        monthly_data = []
        data_start_row = 5
        
        # 2013ë…„ 4ì›”ë¶€í„° ì‹œì‘
        current_year = 2013
        current_month = 4
        
        print(f"\nì›”ë³„ ë°ì´í„° ì½ê¸° ì‹œì‘ ({data_start_row}í–‰ë¶€í„°)...")
        print("ê° í–‰ë³„ ë°ì´í„° í™•ì¸:")
        
        total_months = 0
        consecutive_empty_rows = 0
        max_empty_rows = 5  # ì—°ì† ë¹ˆ í–‰ì´ 5ê°œ ì´ìƒì´ë©´ ì¤‘ë‹¨
        
        for row in range(data_start_row, worksheet.max_row + 1):
            # í˜„ì¬ ë‚ ì§œ ìƒì„± - ë§¤ì›” 1ì¼ë¡œ ì„¤ì •
            current_date = pd.to_datetime(f"{current_year}-{current_month:02d}-01")
            
            # í•´ë‹¹ í–‰ì— ì‹¤ì œ ê°€ê²© ë°ì´í„°ê°€ ìˆëŠ”ì§€ í™•ì¸
            valid_prices_in_row = 0
            row_data = []
            
            for col, region_name in regions:
                cell = worksheet.cell(row=row, column=col)
                if cell.value is not None:
                    try:
                        price_str = str(cell.value).replace(',', '').strip()
                        # ìˆ«ìì¸ì§€ í™•ì¸ (ì†Œìˆ˜ì  í¬í•¨)
                        if price_str and price_str.replace('.', '').replace('-', '').isdigit():
                            price_value = float(price_str)
                            if price_value > 0 and price_value < 100000:  # í˜„ì‹¤ì ì¸ ë²”ìœ„
                                price_per_pyeong = price_value * 3.3  # í‰ë‹¹ ê°€ê²©
                                row_data.append({
                                    'ì‹œì ': current_date,
                                    'ì§€ì—­': region_name,
                                    'ê°€ê²©': price_per_pyeong
                                })
                                valid_prices_in_row += 1
                    except (ValueError, TypeError):
                        continue
            
            # í•´ë‹¹ í–‰ì— ìœ íš¨í•œ ë°ì´í„°ê°€ ìˆìœ¼ë©´ ì¶”ê°€
            if valid_prices_in_row > 0:
                monthly_data.extend(row_data)
                total_months += 1
                consecutive_empty_rows = 0
                
                # ì§„í–‰ìƒí™© ì¶œë ¥ (ë§¤ 12ê°œì›”ë§ˆë‹¤)
                if total_months % 12 == 0:
                    years = total_months // 12
                    print(f"  {years}ë…„ì°¨ ì™„ë£Œ ({current_date.strftime('%Y-%m')}) - í–‰ {row}, ìœ íš¨ ë°ì´í„° {valid_prices_in_row}ê°œ")
                elif total_months <= 5 or total_months % 6 == 0:  # ì²˜ìŒ 5ê°œì›”ê³¼ 6ê°œì›”ë§ˆë‹¤
                    print(f"  {current_date.strftime('%Y-%m')} - í–‰ {row}, ìœ íš¨ ë°ì´í„° {valid_prices_in_row}ê°œ")
            else:
                consecutive_empty_rows += 1
                if consecutive_empty_rows <= 3:  # ì²˜ìŒ ëª‡ ê°œ ë¹ˆ í–‰ë§Œ ì¶œë ¥
                    print(f"  {current_date.strftime('%Y-%m')} - í–‰ {row}, ë°ì´í„° ì—†ìŒ")
            
            # ì—°ì†ìœ¼ë¡œ ë¹ˆ í–‰ì´ ë„ˆë¬´ ë§ìœ¼ë©´ ì¤‘ë‹¨
            if consecutive_empty_rows >= max_empty_rows:
                print(f"ì—°ì† ë¹ˆ í–‰ {consecutive_empty_rows}ê°œ ë°œê²¬ - ë°ì´í„° ëìœ¼ë¡œ íŒë‹¨")
                break
            
            # ë‹¤ìŒ ë‹¬ë¡œ ì´ë™
            current_month += 1
            if current_month > 12:
                current_month = 1
                current_year += 1
        
        workbook.close()
        
        print(f"\nì¶”ì¶œ ê²°ê³¼:")
        print(f"ì²˜ë¦¬ëœ ì›”ìˆ˜: {total_months}ê°œì›”")
        print(f"ì´ ë°ì´í„° í¬ì¸íŠ¸: {len(monthly_data)}ê°œ")
        if total_months > 0:
            last_year = current_year - 1 if current_month == 1 else current_year
            last_month = 12 if current_month == 1 else current_month - 1
            print(f"ê¸°ê°„: 2013-04 ~ {last_year}-{last_month:02d}")
        
        if not monthly_data:
            raise ValueError("ì›”ë³„ ë°ì´í„°ë¥¼ ì¶”ì¶œí•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        
        # DataFrame ìƒì„±
        self.df = pd.DataFrame(monthly_data)
        
        # ì§€ì—­ ëª©ë¡ ìƒì„±
        self.region_list = sorted(list(set([item[1] for item in regions])))
        
        # ì‹œì ìˆœ ì •ë ¬
        self.df = self.df.sort_values(['ì§€ì—­', 'ì‹œì ']).reset_index(drop=True)
        
        print(f"\nìµœì¢… ê²°ê³¼:")
        print(f"DataFrame shape: {self.df.shape}")
        print(f"ì§€ì—­ ê°œìˆ˜: {len(self.region_list)}")
        print(f"ì‹œì  ë²”ìœ„: {self.df['ì‹œì '].min().strftime('%Y-%m')} ~ {self.df['ì‹œì '].max().strftime('%Y-%m')}")
        
        # ì›”ë³„ ë°ì´í„° í™•ì¸
        unique_dates = sorted(self.df['ì‹œì '].unique())
        print(f"ê³ ìœ  ì‹œì  ê°œìˆ˜: {len(unique_dates)}ê°œì›”")
        
        if len(unique_dates) >= 24:
            print(f"ì²« 12ê°œì›”: {[d.strftime('%Y-%m') for d in unique_dates[:12]]}")
            print(f"ë§ˆì§€ë§‰ 12ê°œì›”: {[d.strftime('%Y-%m') for d in unique_dates[-12:]]}")
        else:
            print(f"ëª¨ë“  ì‹œì : {[d.strftime('%Y-%m') for d in unique_dates]}")
        
        # ì‹œì  ê°„ê²© í™•ì¸
        if len(unique_dates) > 1:
            diff = unique_dates[1] - unique_dates[0]
            print(f"ì‹œì  ê°„ê²©: {diff.days}ì¼ (ì›”ë³„ì´ë©´ ì•½ 30ì¼)")
            if 25 <= diff.days <= 35:
                print("âœ“ ì›”ë³„ ë°ì´í„° í™•ì¸ë¨")
            else:
                print("âš  ì›”ë³„ì´ ì•„ë‹ ìˆ˜ ìˆìŒ")
        
        # í•œ ì§€ì—­ì˜ ìƒ˜í”Œ ë°ì´í„° (24ê°œì›”)
        if len(self.df) > 0:
            sample_region = self.region_list[0]
            sample_data = self.df.query('ì§€ì—­ == @sample_region').head(50)  # ë” ë§ì€ ë°ì´í„° í™•ì¸
            print(f"\nìƒ˜í”Œ ë°ì´í„° ({sample_region} ì²« 50ê°œì›”):")
            print("ì‹œì ë³„ ë°ì´í„°:")
            for i, (_, row) in enumerate(sample_data.iterrows()):
                if i < 24:  # ì²« 24ê°œì›” ì¶œë ¥
                    print(f"  {row['ì‹œì '].strftime('%Y-%m')}: {row['ê°€ê²©']:,.0f}ë§Œì›/í‰")
                elif i == 24:
                    print("  ...")
            print(f"ì´ {len(sample_data)}ê°œì›” ë°ì´í„°")
            
            # ì›”ë³„ ë°ì´í„° ì—°ì†ì„± í™•ì¸
            sample_dates = sample_data['ì‹œì '].dt.to_period('M').unique()
            print(f"í•´ë‹¹ ì§€ì—­ ì›”ë³„ ê¸°ê°„: {len(sample_dates)}ê°œì›”")
            if len(sample_dates) >= 10:
                print(f"ì²« 10ê°œì›”: {sample_dates[:10].tolist()}")
                print(f"ë§ˆì§€ë§‰ 10ê°œì›”: {sample_dates[-10:].tolist()}")
        
        # ì „ì²´ ë°ì´í„°ì˜ ì›”ë³„ ë¶„í¬ í™•ì¸
        monthly_counts = self.df.groupby(self.df['ì‹œì '].dt.to_period('M')).size()
        print(f"\nì „ì²´ ì›”ë³„ ë°ì´í„° ë¶„í¬:")
        print(f"ì´ {len(monthly_counts)}ê°œì›”")
        print(f"ê° ì›”ë³„ ì§€ì—­ ìˆ˜: í‰ê·  {monthly_counts.mean():.1f}ê°œ, ìµœì†Œ {monthly_counts.min()}ê°œ, ìµœëŒ€ {monthly_counts.max()}ê°œ")
        
        # price_columns ì„¤ì • (í˜¸í™˜ì„±ì„ ìœ„í•´)
        self.price_columns = ['ê°€ê²©']
        
        print("=" * 60)
        print("ì›”ë³„ ë°ì´í„° ì¶”ì¶œ ì™„ë£Œ!")
        print("=" * 60)
        
        # ë°ì´í„° í™•ì¸ì„ ìœ„í•œ ë””ë²„ê¹… ì½”ë“œ
        print("\n" + "=" * 40)
        print("ë””ë²„ê¹…: ì „ì²´ ë°ì´í„° í™•ì¸")
        print("=" * 40)
        print("ì „ì²´ ë°ì´í„° ìƒ˜í”Œ:")
        print(self.df.head(20))
        print("\nê³ ìœ  ì‹œì ë“¤:")
        unique_times = sorted(self.df['ì‹œì '].unique())
        print(f"ì´ {len(unique_times)}ê°œ ì‹œì ")
        for i, time in enumerate(unique_times[:20]):
            print(f"  {i+1}: {time.strftime('%Y-%m-%d')}")
        
        # ê° ì§€ì—­ë³„ ë°ì´í„° ê°œìˆ˜ í™•ì¸
        region_counts = self.df['ì§€ì—­'].value_counts()
        print(f"\nì§€ì—­ë³„ ë°ì´í„° ê°œìˆ˜:")
        for region, count in region_counts.head(10).items():
            print(f"  {region}: {count}ê°œì›”")
        
        print("=" * 40)
    
    def plot_graph(self):
        """ì‹œê³„ì—´ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°"""
        if self.df is None:
            messagebox.showerror("ì˜¤ë¥˜", "ë¨¼ì € ë°ì´í„°ë¥¼ ë¡œë“œí•´ì£¼ì„¸ìš”.")
            return
        
        mode = self.selection_mode.get()
        selected_regions = []
        
        if mode == "single":
            selected_region = self.region_var.get()
            if not selected_region:
                messagebox.showerror("ì˜¤ë¥˜", "ì§€ì—­ì„ ì„ íƒí•´ì£¼ì„¸ìš”.")
                return
            selected_regions = [selected_region]
        else:
            selected_indices = self.region_listbox.curselection()
            if not selected_indices:
                messagebox.showerror("ì˜¤ë¥˜", "í•˜ë‚˜ ì´ìƒì˜ ì§€ì—­ì„ ì„ íƒí•´ì£¼ì„¸ìš”.")
                return
            selected_regions = [self.region_listbox.get(i) for i in selected_indices]
            
            if len(selected_regions) > 10:
                messagebox.showwarning("ê²½ê³ ", "ì„±ëŠ¥ìƒ ìµœëŒ€ 10ê°œ ì§€ì—­ê¹Œì§€ë§Œ ì„ íƒ ê°€ëŠ¥í•©ë‹ˆë‹¤.")
                selected_regions = selected_regions[:10]
        
        try:
            # ê¸°ì¡´ ê·¸ë˜í”„ ì œê±°
            for widget in self.graph_frame.winfo_children():
                widget.destroy()
                
            # ìƒˆ ê·¸ë˜í”„ ìƒì„±
            fig, ax = plt.subplots(figsize=(14, 8))
            
            # ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì„¤ì •
            colors = plt.cm.tab10(np.linspace(0, 1, len(selected_regions)))
            
            price_col = 'ê°€ê²©'
            chart_type = self.chart_type.get()
            base_point_type = self.base_point.get()
            base_year = self.base_year_var.get() if hasattr(self, 'base_year_var') else None
            all_plot_data = []
            base_dates = []  # ê° ì§€ì—­ì˜ ê¸°ì¤€ì  ë‚ ì§œ ì €ì¥
            
            # ê° ì§€ì—­ë³„ë¡œ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
            for i, region in enumerate(selected_regions):
                # ì§€ì—­ ë°ì´í„° í•„í„°ë§ - query() ì‚¬ìš©
                filtered_data = self.df.query('ì§€ì—­ == @region').copy()
                
                if filtered_data.empty:
                    print(f"ê²½ê³ : {region} ì§€ì—­ì˜ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                    continue
                    
                # ë°ì´í„° ì •ë ¬ ë° ì •ë¦¬
                filtered_data = filtered_data.sort_values('ì‹œì ')
                plot_data = filtered_data.dropna(subset=['ì‹œì ', price_col])
                
                print(f"\n{region} ë°ì´í„° í™•ì¸:")
                print(f"  í•„í„°ë§ëœ ë°ì´í„°: {len(filtered_data)}ê°œ")
                print(f"  ì •ë¦¬ëœ ë°ì´í„°: {len(plot_data)}ê°œ")
                print(f"  ì‹œì  ë²”ìœ„: {plot_data['ì‹œì '].min()} ~ {plot_data['ì‹œì '].max()}")
                
                if plot_data.empty:
                    print(f"ê²½ê³ : {region} ì§€ì—­ì˜ ìœ íš¨í•œ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                    continue
                
                # ì›”ë³„ ë°ì´í„° í™•ì¸
                monthly_data_count = len(plot_data)
                expected_months = (plot_data['ì‹œì '].max() - plot_data['ì‹œì '].min()).days // 30 + 1
                print(f"  ì‹¤ì œ ì›” ìˆ˜: {monthly_data_count}ê°œ, ì˜ˆìƒ ì›” ìˆ˜: {expected_months}ê°œ")
                
                # ì‹œì ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì • (calculate_change_rateì—ì„œ ì‚¬ìš©)
                plot_data_indexed = plot_data.set_index('ì‹œì ')[price_col]
                
                # ì°¨íŠ¸ ìœ í˜•ì— ë”°ë¥¸ ë°ì´í„° ì²˜ë¦¬
                if chart_type == "change_rate":
                    # ìƒìŠ¹/í•˜ë½ë¥  ê³„ì‚°
                    y_data, base_date = self.calculate_change_rate(plot_data_indexed, base_point_type, base_year)
                    y_label = "ìƒìŠ¹/í•˜ë½ë¥  (%)"
                    y_format = lambda x, p: f'{x:+.1f}%'
                    
                    # ê¸°ì¤€ì  ë‚ ì§œ ì €ì¥
                    if base_date:
                        base_dates.append(base_date)
                    
                    # íŠ¹ì • ì—°ë„ ê¸°ì¤€ì¸ ê²½ìš° ê¸°ì¤€ ì—°ë„ ì´í›„ ë°ì´í„°ë§Œ ì‚¬ìš©
                    if base_point_type == "year" and base_date:
                        # plot_dataë„ ê¸°ì¤€ ì—°ë„ ì´í›„ë¡œ í•„í„°ë§
                        plot_data = plot_data[plot_data['ì‹œì '] >= base_date].copy()
                        
                else:
                    # ì ˆëŒ€ê°€ê²©
                    y_data = plot_data[price_col]
                    y_label = "í‰ê· í‰ë‹¨ê°€ (ë§Œì›/í‰)"
                    y_format = lambda x, p: f'{x:,.0f}'
                
                all_plot_data.append((region, plot_data, y_data))
                
                # ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
                color = colors[i]
                if chart_type == "change_rate":
                    # ìƒìŠ¹/í•˜ë½ë¥  ì°¨íŠ¸ì˜ ê²½ìš° ì¸ë±ìŠ¤(ì‹œì ) ì‚¬ìš©
                    line = ax.plot(y_data.index, y_data.values, 
                                 marker='o', linewidth=2, markersize=3, 
                                 label=region, color=color)
                else:
                    # ì ˆëŒ€ê°€ê²© ì°¨íŠ¸
                    line = ax.plot(plot_data['ì‹œì '], y_data, 
                                 marker='o', linewidth=2, markersize=3, 
                                 label=region, color=color)
                
                # ë§ˆì§€ë§‰ ë°ì´í„° í¬ì¸íŠ¸ì— ì§€ì—­ëª…ê³¼ ê°’ í‘œì‹œ
                if len(plot_data) > 0:
                    if chart_type == "change_rate":
                        last_date = y_data.index[-1]
                        last_y = y_data.iloc[-1]
                    else:
                        last_point = plot_data.iloc[-1]
                        last_date = last_point['ì‹œì ']
                        last_y = y_data.iloc[-1]
                    
                    # ì§€ì—­ëª…ì„ ì§§ê²Œ í‘œì‹œ
                    short_name = region.split()[-1] if ' ' in region else region
                    
                    # í•­ìƒ í‰ê· í‰ë‹¨ê°€ë¥¼ í‘œì‹œ
                    last_price = plot_data.iloc[-1][price_col]
                    price_text = f'{last_price:,.0f}'
                    
                    ax.annotate(f'{short_name} {price_text}', 
                               xy=(last_date, last_y),
                               xytext=(5, 0), 
                               textcoords='offset points',
                               bbox=dict(boxstyle='round,pad=0.2', facecolor=color, alpha=0.7),
                               fontsize=8,
                               fontweight='bold',
                               ha='left',
                               va='center',
                               clip_on=False)
            
            if not all_plot_data:
                messagebox.showerror("ì˜¤ë¥˜", "ì„ íƒëœ ì§€ì—­ë“¤ì— ìœ íš¨í•œ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ìƒìŠ¹/í•˜ë½ë¥  ì°¨íŠ¸ì¸ ê²½ìš° 0% ê¸°ì¤€ì„  ì¶”ê°€
            if chart_type == "change_rate":
                ax.axhline(y=0, color='gray', linestyle='--', alpha=0.7, linewidth=1.5, label='0% ê¸°ì¤€ì„ ')
                
                # íŠ¹ì • ì—°ë„ ê¸°ì¤€ì¸ ê²½ìš° ê¸°ì¤€ì ì— ìˆ˜ì§ì„  ì¶”ê°€
                if base_point_type == "year" and base_dates:
                    # ëª¨ë“  ì§€ì—­ì˜ ê¸°ì¤€ì ì´ ê°™ë‹¤ê³  ê°€ì •í•˜ê³  ì²« ë²ˆì§¸ ì‚¬ìš©
                    base_date = base_dates[0]
                    ax.axvline(x=base_date, color='red', linestyle=':', alpha=0.7, linewidth=1.5, 
                              label=f'ê¸°ì¤€ì  ({base_date.strftime("%Y-%m")})')
            
            # ê·¸ë˜í”„ ì„¤ì •
            if len(selected_regions) == 1:
                if chart_type == "change_rate":
                    if base_point_type == "first":
                        base_text = "ì²« ë°ì´í„° ëŒ€ë¹„"
                    elif base_point_type == "previous":
                        base_text = "ì´ì „ ê¸°ê°„ ëŒ€ë¹„"
                    else:
                        base_text = f"{base_year}ë…„ ê¸°ì¤€"
                    title = f'{selected_regions[0]} ìƒìŠ¹/í•˜ë½ë¥  ({base_text})'
                else:
                    title = f'{selected_regions[0]} í‰ê· í‰ë‹¨ê°€'
            else:
                if chart_type == "change_rate":
                    if base_point_type == "first":
                        base_text = "ì²« ë°ì´í„° ëŒ€ë¹„"
                    elif base_point_type == "previous":
                        base_text = "ì´ì „ ê¸°ê°„ ëŒ€ë¹„"
                    else:
                        base_text = f"{base_year}ë…„ ê¸°ì¤€"
                    title = f'ì§€ì—­ë³„ ìƒìŠ¹/í•˜ë½ë¥  ë¹„êµ ({base_text}, {len(selected_regions)}ê°œ ì§€ì—­)'
                else:
                    title = f'ì§€ì—­ë³„ í‰ê· í‰ë‹¨ê°€ ë¹„êµ ({len(selected_regions)}ê°œ ì§€ì—­)'
            
            ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
            ax.set_xlabel('ì‹œì ', fontsize=12)
            ax.set_ylabel(y_label, fontsize=12)
            ax.grid(True, alpha=0.3)
            
            # yì¶• í¬ë§· ì„¤ì •
            ax.yaxis.set_major_formatter(plt.FuncFormatter(y_format))
            
            # xì¶• ë‚ ì§œ í¬ë§· ì„¤ì • - ëª¨ë“  ì›”ë³„ ë°ì´í„° í‘œì‹œ
            import matplotlib.dates as mdates
            
            # ëª¨ë“  ì›”ë³„ ë°ì´í„°ê°€ í‘œì‹œë˜ë„ë¡ ì„¤ì •
            ax.xaxis.set_major_locator(mdates.YearLocator())  # ì—°ë„ë³„ major tick
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))  # ì—°ë„ í‘œì‹œ
            ax.xaxis.set_minor_locator(mdates.MonthLocator())  # ëª¨ë“  ì›” minor tick
            
            # ë°ì´í„° í¬ì¸íŠ¸ í™•ì¸ì„ ìœ„í•œ ë””ë²„ê¹…
            if len(all_plot_data) > 0:
                sample_region, sample_data, sample_y = all_plot_data[0]
                print(f"\nê·¸ë˜í”„ ë°ì´í„° í™•ì¸ ({sample_region}):")
                print(f"ì´ ë°ì´í„° í¬ì¸íŠ¸: {len(sample_data)}ê°œ")
                print(f"ì‹œì  ë²”ìœ„: {sample_data['ì‹œì '].min()} ~ {sample_data['ì‹œì '].max()}")
                print(f"ì²« 10ê°œ ì‹œì : {sample_data['ì‹œì '].head(10).tolist()}")
            
            # ë‚ ì§œ ë¼ë²¨ íšŒì „
            fig.autofmt_xdate(rotation=45)
            
            # ë²”ë¡€ ì„¤ì •
            if len(selected_regions) > 1 or chart_type == "change_rate":
                ax.legend(loc='upper left', bbox_to_anchor=(0, 1), ncol=min(3, len(selected_regions)))
            
            # ì—¬ë°± ì¡°ì • - ì˜¤ë¥¸ìª½ ì—¬ë°±ì„ ì¶©ë¶„íˆ í™•ë³´
            plt.subplots_adjust(left=0.1, right=0.85, top=0.9, bottom=0.15)
            
            # ê·¸ë˜í”„ë¥¼ GUIì— ì¶”ê°€
            canvas = FigureCanvasTkAgg(fig, master=self.graph_frame)
            canvas.draw()
            canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
            
            # í†µê³„ ì •ë³´ í‘œì‹œ
            if all_plot_data:
                if chart_type == "change_rate":
                    self.show_change_rate_statistics(all_plot_data, base_point_type)
                else:
                    first_region, first_data, first_y = all_plot_data[0]
                    self.show_statistics(first_data[price_col], first_region)
            
            # ì„ íƒëœ ì§€ì—­ ì €ì¥
            self.save_settings()
            
        except Exception as e:
            messagebox.showerror("ì˜¤ë¥˜", f"ê·¸ë˜í”„ ìƒì„± ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {str(e)}")
            import traceback
            print(f"ìƒì„¸ ì˜¤ë¥˜: {traceback.format_exc()}")


    
    def select_file(self):
        file_path = filedialog.askopenfilename(
            title="ì—‘ì…€ íŒŒì¼ ì„ íƒ",
            filetypes=[("Excel files", "*.xlsx *.xls"), ("All files", "*.*")]
        )
        if file_path:
            self.file_path_var.set(file_path)
            
    def load_data(self):
        file_path = self.file_path_var.get()
        if not file_path:
            messagebox.showerror("ì˜¤ë¥˜", "ë¨¼ì € ì—‘ì…€ íŒŒì¼ì„ ì„ íƒí•´ì£¼ì„¸ìš”.")
            return
            
        try:
            # íŒŒì¼ ê²½ë¡œ ì •ê·œí™” ë° ê¶Œí•œ í™•ì¸
            import os
            if not os.path.exists(file_path):
                messagebox.showerror("ì˜¤ë¥˜", "íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤. íŒŒì¼ ê²½ë¡œë¥¼ í™•ì¸í•´ì£¼ì„¸ìš”.")
                return
                
            if not os.access(file_path, os.R_OK):
                messagebox.showerror("ì˜¤ë¥˜", "íŒŒì¼ ì½ê¸° ê¶Œí•œì´ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ì—‘ì…€ íŒŒì¼ì˜ ì‹œíŠ¸ ëª©ë¡ í™•ì¸
            try:
                xl_file = pd.ExcelFile(file_path)
                sheet_names = xl_file.sheet_names
                print(f"ì‚¬ìš© ê°€ëŠ¥í•œ ì‹œíŠ¸: {sheet_names}")
            except Exception as e:
                messagebox.showerror("ì˜¤ë¥˜", f"ì—‘ì…€ íŒŒì¼ì„ ì—´ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {str(e)}")
                return
            
            # '47.ã¡ë‹¹ì•„íŒŒíŠ¸í‰ê· ë§¤ë§¤' ì‹œíŠ¸ í™•ì¸
            target_sheet = None
            for sheet in sheet_names:
                if '47' in sheet and 'ã¡' in sheet and 'ì•„íŒŒíŠ¸' in sheet:
                    target_sheet = sheet
                    break
                    
            if target_sheet is None:
                # ì‹œíŠ¸ê°€ ì—†ìœ¼ë©´ ì‚¬ìš©ìì—ê²Œ ì„ íƒí•˜ê²Œ í•¨
                sheet_selection = self.select_sheet(sheet_names)
                if sheet_selection:
                    target_sheet = sheet_selection
                else:
                    return
            
            # openpyxlë¡œ ì§ì ‘ ì›”ë³„ ë°ì´í„° ì½ê¸°
            print("ì›”ë³„ ë°ì´í„° ì§ì ‘ ì¶”ì¶œ ì‹œì‘...")
            self.read_monthly_data_with_openpyxl(file_path, target_sheet)
            
            # ì§€ì—­ ì½¤ë³´ë°•ìŠ¤ ì—…ë°ì´íŠ¸
            self.update_region_combos()
            
            messagebox.showinfo("ì™„ë£Œ", f"ì›”ë³„ ë°ì´í„° ë¡œë“œ ì™„ë£Œ!\nì‹œíŠ¸: {target_sheet}\nì´ {len(self.df)}ê°œ ì›”ë³„ ë°ì´í„°")
            
            # íŒŒì¼ ê²½ë¡œ ì €ì¥
            self.save_settings()
            
        except Exception as e:
            messagebox.showerror("ì˜¤ë¥˜", f"íŒŒì¼ ì½ê¸° ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {str(e)}")
    
    def select_sheet(self, sheet_names):
        """ì‹œíŠ¸ ì„ íƒ ë‹¤ì´ì–¼ë¡œê·¸"""
        sheet_window = tk.Toplevel(self.root)
        sheet_window.title("ì‹œíŠ¸ ì„ íƒ")
        sheet_window.geometry("400x300")
        sheet_window.grab_set()
        
        ttk.Label(sheet_window, text="ë¶„ì„í•  ì‹œíŠ¸ë¥¼ ì„ íƒí•´ì£¼ì„¸ìš”:").pack(pady=10)
        
        selected_sheet = tk.StringVar()
        
        # ì‹œíŠ¸ ëª©ë¡ í‘œì‹œ
        listbox = tk.Listbox(sheet_window, height=10)
        for sheet in sheet_names:
            listbox.insert(tk.END, sheet)
        listbox.pack(pady=10, padx=20, fill=tk.BOTH, expand=True)
        
        def on_select():
            selection = listbox.curselection()
            if selection:
                selected_sheet.set(sheet_names[selection[0]])
                sheet_window.destroy()
        
        def on_cancel():
            selected_sheet.set("")
            sheet_window.destroy()
        
        button_frame = ttk.Frame(sheet_window)
        button_frame.pack(pady=10)
        
        ttk.Button(button_frame, text="ì„ íƒ", command=on_select).pack(side=tk.LEFT, padx=5)
        ttk.Button(button_frame, text="ì·¨ì†Œ", command=on_cancel).pack(side=tk.LEFT, padx=5)
        
        # ë”ë¸”í´ë¦­ìœ¼ë¡œë„ ì„ íƒ ê°€ëŠ¥
        listbox.bind('<Double-1>', lambda e: on_select())
        
        sheet_window.wait_window()
        return selected_sheet.get() if selected_sheet.get() else None
            
    def preprocess_data(self):
        """ë°ì´í„° ì „ì²˜ë¦¬"""
        print("ì›ë³¸ ë°ì´í„° êµ¬ì¡°:")
        print(f"Shape: {self.df.shape}")
        print(f"ì²« 5í–‰:\n{self.df.head()}")
        
        if len(self.df) < 4:
            raise ValueError("ë°ì´í„°ê°€ ë¶€ì¡±í•©ë‹ˆë‹¤. ìµœì†Œ 4í–‰ì´ í•„ìš”í•©ë‹ˆë‹¤.")
        
        # ì›ë³¸ ì—‘ì…€ íŒŒì¼ì—ì„œ ë³‘í•© ì •ë³´ ì½ê¸°
        file_path = self.file_path_var.get()
        region_mapping = self.get_merged_cell_mapping(file_path)
        
        # ì‹¤ì œ ë°ì´í„°ëŠ” 4í–‰ë¶€í„° (ì¸ë±ìŠ¤ 3ë¶€í„°)
        data_df = self.df.iloc[3:].copy()
        data_df.reset_index(drop=True, inplace=True)
        
        # ì²« ë²ˆì§¸ ì»¬ëŸ¼ì€ ì‹œì  ì •ë³´
        time_column = data_df.iloc[:, 0]
        
        # ì§€ì—­ë³„ ë°ì´í„° êµ¬ì¡° ë§Œë“¤ê¸°
        processed_data = []
        region_list = []
        
        for col_idx, (sido_clean, sigungu_clean) in region_mapping.items():
            # ì§€ì—­ëª… ì¡°í•© (ì‹œë„ + ì‹œêµ°êµ¬)
            region_name = f"{sido_clean} {sigungu_clean}"
            region_list.append(region_name)
            
            # í•´ë‹¹ ì§€ì—­ì˜ ê°€ê²© ë°ì´í„°
            if col_idx < len(data_df.columns):
                price_data = data_df.iloc[:, col_idx]
                
                # ê° ì‹œì ë³„ë¡œ ë°ì´í„° ìƒì„±
                for time_idx, (time_val, price_val) in enumerate(zip(time_column, price_data)):
                    # ì‹œì  ë°ì´í„° ì •ë¦¬
                    if pd.isna(time_val) or str(time_val).strip() == '':
                        continue
                    
                    # ê°€ê²© ë°ì´í„° ì •ë¦¬
                    if pd.isna(price_val):
                        continue
                        
                    try:
                        # ê°€ê²© ë°ì´í„°ë¥¼ ìˆ«ìë¡œ ë³€í™˜í•˜ê³  3.3ì„ ê³±í•´ì„œ í‰ë‹¹ ê°€ê²©ìœ¼ë¡œ ë³€í™˜
                        price_clean = str(price_val).replace(',', '').replace(' ', '').strip()
                        if price_clean == '' or price_clean == 'nan':
                            continue
                        price_numeric = float(price_clean) * 3.3  # 1í‰(3.3ã¡) ê¸°ì¤€ìœ¼ë¡œ ë³€í™˜
                        
                        processed_data.append({
                            'ì‹œì ': time_val,
                            'ì§€ì—­': region_name,
                            'ê°€ê²©': price_numeric
                        })
                    except (ValueError, TypeError):
                        continue
        
        if not processed_data:
            raise ValueError("ì²˜ë¦¬í•  ìˆ˜ ìˆëŠ” ìœ íš¨í•œ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
        
        # ìƒˆë¡œìš´ DataFrame ìƒì„±
        self.df = pd.DataFrame(processed_data)
        
        # ì§€ì—­ ëª©ë¡ ì €ì¥
        self.region_list = region_list
        
        print(f"ì²˜ë¦¬ëœ ë°ì´í„° shape: {self.df.shape}")
        print(f"ì§€ì—­ ëª©ë¡: {region_list}")
        print(f"ìƒ˜í”Œ ë°ì´í„°:\n{self.df.head()}")
        
        # ì‹œì  ë°ì´í„° ì²˜ë¦¬
        def parse_date(date_val):
            if pd.isna(date_val):
                return None
                
            date_str = str(date_val).strip()
            
            # 2013.4 í˜•ì‹ ì²˜ë¦¬
            if '.' in date_str and len(date_str.split('.')) == 2:
                try:
                    year, month = date_str.split('.')
                    return pd.to_datetime(f"{year}-{month.zfill(2)}-01")
                except:
                    pass
            
            # ë‹¤ë¥¸ í˜•ì‹ë“¤ ì‹œë„
            formats_to_try = [
                '%Y.%m', '%Y-%m', '%Y/%m',
                '%Y.%m.%d', '%Y-%m-%d', '%Y/%m/%d',
                '%Y%m',
            ]
            
            for fmt in formats_to_try:
                try:
                    return pd.to_datetime(date_str, format=fmt)
                except:
                    continue
            
            try:
                return pd.to_datetime(date_str, errors='coerce')
            except:
                return None
        
        self.df['ì‹œì '] = self.df['ì‹œì '].apply(parse_date)
        
        # ìœ íš¨í•œ ì‹œì  ë°ì´í„°ë§Œ ìœ ì§€
        self.df = self.df.dropna(subset=['ì‹œì ']).reset_index(drop=True)
        
        # ì‹œì ìˆœ ì •ë ¬
        self.df = self.df.sort_values('ì‹œì ').reset_index(drop=True)
        
        print(f"ìµœì¢… ì²˜ë¦¬ëœ ë°ì´í„° shape: {self.df.shape}")
        print(f"ì‹œì  ë²”ìœ„: {self.df['ì‹œì '].min()} ~ {self.df['ì‹œì '].max()}")
        
        # price_columns ì„¤ì • (í˜¸í™˜ì„±ì„ ìœ„í•´)
        self.price_columns = ['ê°€ê²©']
        
    def get_merged_cell_mapping(self, file_path):
        """ì—‘ì…€ íŒŒì¼ì—ì„œ ë³‘í•©ëœ ì…€ ì •ë³´ë¥¼ ì½ì–´ì„œ ì§€ì—­ ë§¤í•‘ ìƒì„±"""
        region_mapping = {}
        
        try:
            print("ì—‘ì…€ íŒŒì¼ ì½ëŠ” ì¤‘...")
            
            # openpyxlë¡œ ì—‘ì…€ íŒŒì¼ ì—´ê¸°
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                workbook = openpyxl.load_workbook(file_path, data_only=True)
            
            # ì‹œíŠ¸ ì„ íƒ
            sheet_name = None
            for name in workbook.sheetnames:
                if '47' in name and 'ã¡' in name and 'ì•„íŒŒíŠ¸' in name:
                    sheet_name = name
                    break
            
            if sheet_name is None:
                sheet_name = workbook.sheetnames[0]
            
            worksheet = workbook[sheet_name]
            print(f"ì‹œíŠ¸: {sheet_name}")
            
            # ëª¨ë“  ì»¬ëŸ¼ì— ëŒ€í•´ ì§€ì—­ëª… ë§¤í•‘ ìƒì„±
            sido_mapping = {}
            sigungu_mapping = {}
            
            # 1ë‹¨ê³„: ëª¨ë“  ì…€ ê°’ ì½ê¸°
            for col in range(1, worksheet.max_column + 1):
                # 2í–‰ (ì‹œ/ë„)
                cell_2 = worksheet.cell(row=2, column=col)
                if cell_2.value and str(cell_2.value).strip():
                    value = str(cell_2.value).strip()
                    sido_mapping[col] = value
                
                # 3í–‰ (ì‹œ/êµ°/êµ¬)
                cell_3 = worksheet.cell(row=3, column=col)
                if cell_3.value and str(cell_3.value).strip():
                    value = str(cell_3.value).strip()
                    sigungu_mapping[col] = value
            
            # 2ë‹¨ê³„: ë³‘í•©ëœ ì…€ ì²˜ë¦¬
            for merged_range in worksheet.merged_cells.ranges:
                min_row, max_row = merged_range.min_row, merged_range.max_row
                min_col, max_col = merged_range.min_col, merged_range.max_col
                
                # 2í–‰ ë³‘í•©
                if min_row <= 2 <= max_row:
                    master_cell = worksheet.cell(row=2, column=min_col)
                    if master_cell.value and str(master_cell.value).strip():
                        merged_value = str(master_cell.value).strip()
                        for col in range(min_col, max_col + 1):
                            sido_mapping[col] = merged_value
                
                # 3í–‰ ë³‘í•©
                if min_row <= 3 <= max_row:
                    master_cell = worksheet.cell(row=3, column=min_col)
                    if master_cell.value and str(master_cell.value).strip():
                        merged_value = str(master_cell.value).strip()
                        for col in range(min_col, max_col + 1):
                            sigungu_mapping[col] = merged_value
                
                # 2-3í–‰ í•¨ê»˜ ë³‘í•©
                if min_row <= 2 and max_row >= 3:
                    master_cell = worksheet.cell(row=min_row, column=min_col)
                    if master_cell.value and str(master_cell.value).strip():
                        region_value = str(master_cell.value).strip()
                        
                        if ' ' in region_value:
                            parts = region_value.split(' ', 1)
                            for col in range(min_col, max_col + 1):
                                sido_mapping[col] = parts[0]
                                sigungu_mapping[col] = parts[1]
                        else:
                            for col in range(min_col, max_col + 1):
                                sido_mapping[col] = region_value
                                sigungu_mapping[col] = region_value
            
            # 3ë‹¨ê³„: ë¹ˆ ì…€ ì±„ìš°ê¸° (ì‹œ/ë„ë§Œ)
            current_sido = None
            for col in range(1, worksheet.max_column + 1):
                if col in sido_mapping and sido_mapping[col]:
                    current_sido = sido_mapping[col]
                elif current_sido and col not in sido_mapping:
                    sido_mapping[col] = current_sido
            
            # 4ë‹¨ê³„: ìœ íš¨í•œ ì§€ì—­ë§Œ ë§¤í•‘ (Bì—´ë¶€í„°)
            exclude_terms = ['êµ¬ë¶„', 'ì§€ì—­', 'ì‹œì ', 'ë…„ë„']
            
            for col in range(2, worksheet.max_column + 1):  # Bì—´ë¶€í„°
                pandas_idx = col - 1
                
                sido = sido_mapping.get(col, '').strip()
                sigungu = sigungu_mapping.get(col, '').strip()
                
                # ì‹œ/ë„ë§Œ ìˆê³  ì‹œ/êµ°/êµ¬ê°€ ì—†ëŠ” ê²½ìš° ì²˜ë¦¬
                if sido and not sigungu:
                    if sido in ['ì „êµ­', 'ì„œìš¸íŠ¹ë³„ì‹œ'] or 'Total' in sido or 'Seoul' in sido:
                        sigungu = sido
                
                # ì œì™¸í•  ìš©ì–´ ì²´í¬
                if sido in exclude_terms or sigungu in exclude_terms:
                    continue
                
                # ìœ íš¨í•œ ì§€ì—­ëª…ì´ ìˆìœ¼ë©´ ë§¤í•‘
                if sido and sigungu:
                    region_mapping[pandas_idx] = (sido, sigungu)
            
            workbook.close()
            print(f"ë§¤í•‘ ì™„ë£Œ: {len(region_mapping)}ê°œ ì§€ì—­")
            
        except Exception as e:
            print(f"openpyxl ì²˜ë¦¬ ì‹¤íŒ¨: {e}")
            return self.get_basic_region_mapping()
        
        return region_mapping
    
    def get_basic_region_mapping(self):
        """ê¸°ë³¸ ë°©ì‹ìœ¼ë¡œ ì§€ì—­ ë§¤í•‘ ìƒì„±"""
        region_mapping = {}
        
        # 2í–‰ê³¼ 3í–‰ì—ì„œ ì§€ì—­ ì •ë³´ ì¶”ì¶œ
        sido_row = self.df.iloc[1]
        sigungu_row = self.df.iloc[2]
        
        exclude_terms = ['êµ¬ë¶„', 'ì§€ì—­', 'ì‹œì ', 'ë…„ë„']
        
        current_sido = None
        for col_idx in range(1, len(self.df.columns)):
            if col_idx == 0:
                continue
                
            # ì‹œ/ë„ ì°¾ê¸°
            sido_val = sido_row.iloc[col_idx] if col_idx < len(sido_row) else None
            if pd.notna(sido_val) and str(sido_val).strip():
                potential_sido = str(sido_val).strip()
                if potential_sido not in exclude_terms:
                    current_sido = potential_sido
            
            # ì‹œ/êµ°/êµ¬ ì°¾ê¸°
            sigungu_val = sigungu_row.iloc[col_idx] if col_idx < len(sigungu_row) else None
            sigungu = None
            
            if pd.notna(sigungu_val) and str(sigungu_val).strip():
                sigungu = str(sigungu_val).strip()
            elif current_sido and current_sido in ['ì „êµ­', 'ì„œìš¸íŠ¹ë³„ì‹œ']:
                sigungu = current_sido
            
            # ì œì™¸í•  ìš©ì–´ ì²´í¬
            if sigungu and sigungu in exclude_terms:
                continue
            
            if current_sido and sigungu and current_sido not in exclude_terms:
                region_mapping[col_idx] = (current_sido, sigungu)
        
        return region_mapping
        
    def update_region_combos(self):
        """ì§€ì—­ ì½¤ë³´ë°•ìŠ¤ì™€ ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ ì—…ë°ì´íŠ¸"""
        if self.df is None or not hasattr(self, 'region_list'):
            return
            
        self.region_combo['values'] = self.region_list
        
        self.region_listbox.delete(0, tk.END)
        for region in self.region_list:
            self.region_listbox.insert(tk.END, region)
        
        self.region_var.set('')
        self.selected_regions_var.set("ì—†ìŒ")
                    
    def plot_graph(self):
        """ì‹œê³„ì—´ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°"""
        if self.df is None:
            messagebox.showerror("ì˜¤ë¥˜", "ë¨¼ì € ë°ì´í„°ë¥¼ ë¡œë“œí•´ì£¼ì„¸ìš”.")
            return
        
        mode = self.selection_mode.get()
        selected_regions = []
        
        if mode == "single":
            selected_region = self.region_var.get()
            if not selected_region:
                messagebox.showerror("ì˜¤ë¥˜", "ì§€ì—­ì„ ì„ íƒí•´ì£¼ì„¸ìš”.")
                return
            selected_regions = [selected_region]
        else:
            selected_indices = self.region_listbox.curselection()
            if not selected_indices:
                messagebox.showerror("ì˜¤ë¥˜", "í•˜ë‚˜ ì´ìƒì˜ ì§€ì—­ì„ ì„ íƒí•´ì£¼ì„¸ìš”.")
                return
            selected_regions = [self.region_listbox.get(i) for i in selected_indices]
            
            if len(selected_regions) > 10:
                messagebox.showwarning("ê²½ê³ ", "ì„±ëŠ¥ìƒ ìµœëŒ€ 10ê°œ ì§€ì—­ê¹Œì§€ë§Œ ì„ íƒ ê°€ëŠ¥í•©ë‹ˆë‹¤.")
                selected_regions = selected_regions[:10]
        
        try:
            # ê¸°ì¡´ ê·¸ë˜í”„ ì œê±°
            for widget in self.graph_frame.winfo_children():
                widget.destroy()
                
            # ìƒˆ ê·¸ë˜í”„ ìƒì„±
            fig, ax = plt.subplots(figsize=(14, 8))
            
            # ìƒ‰ìƒ íŒ”ë ˆíŠ¸ ì„¤ì •
            colors = plt.cm.tab10(np.linspace(0, 1, len(selected_regions)))
            
            price_col = 'ê°€ê²©'
            chart_type = self.chart_type.get()
            base_point_type = self.base_point.get()
            base_year = self.base_year_var.get() if hasattr(self, 'base_year_var') else None
            all_plot_data = []
            base_dates = []  # ê° ì§€ì—­ì˜ ê¸°ì¤€ì  ë‚ ì§œ ì €ì¥
            
            # ê° ì§€ì—­ë³„ë¡œ ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
            for i, region in enumerate(selected_regions):
                # ì§€ì—­ ë°ì´í„° í•„í„°ë§
                mask = self.df['ì§€ì—­'] == region
                filtered_data = self.df[mask].copy()
                
                if filtered_data.empty:
                    continue
                    
                # ë°ì´í„° ì •ë ¬ ë° ì •ë¦¬
                filtered_data = filtered_data.sort_values('ì‹œì ')
                plot_data = filtered_data.dropna(subset=['ì‹œì ', price_col])
                
                if plot_data.empty:
                    continue
                
                # ì‹œì ì„ ì¸ë±ìŠ¤ë¡œ ì„¤ì • (calculate_change_rateì—ì„œ ì‚¬ìš©)
                plot_data_indexed = plot_data.set_index('ì‹œì ')[price_col]
                
                # ì°¨íŠ¸ ìœ í˜•ì— ë”°ë¥¸ ë°ì´í„° ì²˜ë¦¬
                if chart_type == "change_rate":
                    # ìƒìŠ¹/í•˜ë½ë¥  ê³„ì‚°
                    y_data, base_date = self.calculate_change_rate(plot_data_indexed, base_point_type, base_year)
                    y_label = "ìƒìŠ¹/í•˜ë½ë¥  (%)"
                    y_format = lambda x, p: f'{x:+.1f}%'
                    
                    # ê¸°ì¤€ì  ë‚ ì§œ ì €ì¥
                    if base_date:
                        base_dates.append(base_date)
                    
                    # íŠ¹ì • ì—°ë„ ê¸°ì¤€ì¸ ê²½ìš° ê¸°ì¤€ ì—°ë„ ì´í›„ ë°ì´í„°ë§Œ ì‚¬ìš©
                    if base_point_type == "year" and base_date:
                        # plot_dataë„ ê¸°ì¤€ ì—°ë„ ì´í›„ë¡œ í•„í„°ë§
                        plot_data = plot_data[plot_data['ì‹œì '] >= base_date].copy()
                        
                else:
                    # ì ˆëŒ€ê°€ê²©
                    y_data = plot_data[price_col]
                    y_label = "í‰ê· í‰ë‹¨ê°€ (ë§Œì›/í‰)"
                    y_format = lambda x, p: f'{x:,.0f}'
                
                all_plot_data.append((region, plot_data, y_data))
                
                # ì„  ê·¸ë˜í”„ ê·¸ë¦¬ê¸°
                color = colors[i]
                if chart_type == "change_rate":
                    # ìƒìŠ¹/í•˜ë½ë¥  ì°¨íŠ¸ì˜ ê²½ìš° ì¸ë±ìŠ¤(ì‹œì ) ì‚¬ìš©
                    line = ax.plot(y_data.index, y_data.values, 
                                 marker='o', linewidth=2, markersize=3, 
                                 label=region, color=color)
                else:
                    # ì ˆëŒ€ê°€ê²© ì°¨íŠ¸
                    line = ax.plot(plot_data['ì‹œì '], y_data, 
                                 marker='o', linewidth=2, markersize=3, 
                                 label=region, color=color)
                
                # ë§ˆì§€ë§‰ ë°ì´í„° í¬ì¸íŠ¸ì— ì§€ì—­ëª…ê³¼ ê°’ í‘œì‹œ
                if len(plot_data) > 0:
                    if chart_type == "change_rate":
                        last_date = y_data.index[-1]
                        last_y = y_data.iloc[-1]
                    else:
                        last_point = plot_data.iloc[-1]
                        last_date = last_point['ì‹œì ']
                        last_y = y_data.iloc[-1]
                    
                    # ì§€ì—­ëª…ì„ ì§§ê²Œ í‘œì‹œ
                    short_name = region.split()[-1] if ' ' in region else region
                    
                    # í•­ìƒ í‰ê· í‰ë‹¨ê°€ë¥¼ í‘œì‹œ
                    last_price = plot_data.iloc[-1][price_col]
                    price_text = f'{last_price:,.0f}'
                    
                    ax.annotate(f'{short_name} {price_text}', 
                               xy=(last_date, last_y),
                               xytext=(5, 0), 
                               textcoords='offset points',
                               bbox=dict(boxstyle='round,pad=0.2', facecolor=color, alpha=0.7),
                               fontsize=8,
                               fontweight='bold',
                               ha='left',
                               va='center',
                               clip_on=False)  # í´ë¦¬í•‘ ë°©ì§€
            
            if not all_plot_data:
                messagebox.showerror("ì˜¤ë¥˜", "ì„ íƒëœ ì§€ì—­ë“¤ì— ìœ íš¨í•œ ë°ì´í„°ê°€ ì—†ìŠµë‹ˆë‹¤.")
                return
            
            # ìƒìŠ¹/í•˜ë½ë¥  ì°¨íŠ¸ì¸ ê²½ìš° 0% ê¸°ì¤€ì„  ì¶”ê°€
            if chart_type == "change_rate":
                ax.axhline(y=0, color='gray', linestyle='--', alpha=0.7, linewidth=1.5, label='0% ê¸°ì¤€ì„ ')
                
                # íŠ¹ì • ì—°ë„ ê¸°ì¤€ì¸ ê²½ìš° ê¸°ì¤€ì ì— ìˆ˜ì§ì„  ì¶”ê°€
                if base_point_type == "year" and base_dates:
                    # ëª¨ë“  ì§€ì—­ì˜ ê¸°ì¤€ì ì´ ê°™ë‹¤ê³  ê°€ì •í•˜ê³  ì²« ë²ˆì§¸ ì‚¬ìš©
                    base_date = base_dates[0]
                    ax.axvline(x=base_date, color='red', linestyle=':', alpha=0.7, linewidth=1.5, 
                              label=f'ê¸°ì¤€ì  ({base_date.strftime("%Y-%m")})')
            
            # ê·¸ë˜í”„ ì„¤ì •
            if len(selected_regions) == 1:
                if chart_type == "change_rate":
                    if base_point_type == "first":
                        base_text = "ì²« ë°ì´í„° ëŒ€ë¹„"
                    elif base_point_type == "previous":
                        base_text = "ì´ì „ ê¸°ê°„ ëŒ€ë¹„"
                    else:
                        base_text = f"{base_year}ë…„ ê¸°ì¤€"
                    title = f'{selected_regions[0]} ìƒìŠ¹/í•˜ë½ë¥  ({base_text})'
                else:
                    title = f'{selected_regions[0]} í‰ê· í‰ë‹¨ê°€'
            else:
                if chart_type == "change_rate":
                    if base_point_type == "first":
                        base_text = "ì²« ë°ì´í„° ëŒ€ë¹„"
                    elif base_point_type == "previous":
                        base_text = "ì´ì „ ê¸°ê°„ ëŒ€ë¹„"
                    else:
                        base_text = f"{base_year}ë…„ ê¸°ì¤€"
                    title = f'ì§€ì—­ë³„ ìƒìŠ¹/í•˜ë½ë¥  ë¹„êµ ({base_text}, {len(selected_regions)}ê°œ ì§€ì—­)'
                else:
                    title = f'ì§€ì—­ë³„ í‰ê· í‰ë‹¨ê°€ ë¹„êµ ({len(selected_regions)}ê°œ ì§€ì—­)'
            
            ax.set_title(title, fontsize=16, fontweight='bold', pad=20)
            ax.set_xlabel('ì‹œì ', fontsize=12)
            ax.set_ylabel(y_label, fontsize=12)
            ax.grid(True, alpha=0.3)
            
            # yì¶• í¬ë§· ì„¤ì •
            ax.yaxis.set_major_formatter(plt.FuncFormatter(y_format))
            
            # xì¶• ë‚ ì§œ í¬ë§· ì„¤ì •
            fig.autofmt_xdate()
            
            # ë²”ë¡€ ì„¤ì •
            if len(selected_regions) > 1 or chart_type == "change_rate":
                ax.legend(loc='upper left', bbox_to_anchor=(0, 1), ncol=min(3, len(selected_regions)))
            
            # ì—¬ë°± ì¡°ì • - ì˜¤ë¥¸ìª½ ì—¬ë°±ì„ ì¶©ë¶„íˆ í™•ë³´
            plt.subplots_adjust(left=0.1, right=0.85, top=0.9, bottom=0.15)
            
            # ê·¸ë˜í”„ë¥¼ GUIì— ì¶”ê°€
            canvas = FigureCanvasTkAgg(fig, master=self.graph_frame)
            canvas.draw()
            canvas.get_tk_widget().pack(fill=tk.BOTH, expand=True)
            
            # í†µê³„ ì •ë³´ í‘œì‹œ
            if all_plot_data:
                if chart_type == "change_rate":
                    self.show_change_rate_statistics(all_plot_data, base_point_type)
                else:
                    first_region, first_data, first_y = all_plot_data[0]
                    self.show_statistics(first_data[price_col], first_region)
            
            # ì„ íƒëœ ì§€ì—­ ì €ì¥
            self.save_settings()
            
        except Exception as e:
            messagebox.showerror("ì˜¤ë¥˜", f"ê·¸ë˜í”„ ìƒì„± ì¤‘ ì˜¤ë¥˜ê°€ ë°œìƒí–ˆìŠµë‹ˆë‹¤: {str(e)}")
    
    def show_change_rate_statistics(self, all_plot_data, base_point_type):
        """ìƒìŠ¹/í•˜ë½ë¥  í†µê³„ ì •ë³´ í‘œì‹œ"""
        if base_point_type == "first":
            base_text = "ì²« ë°ì´í„° ëŒ€ë¹„"
        elif base_point_type == "previous":
            base_text = "ì´ì „ ê¸°ê°„ ëŒ€ë¹„"
        else:
            base_year = self.base_year_var.get() if hasattr(self, 'base_year_var') else '2020'
            base_text = f"{base_year}ë…„ ê¸°ì¤€"
        
        if len(all_plot_data) == 1:
            # ë‹¨ì¼ ì§€ì—­
            region, data, change_rates = all_plot_data[0]
            if hasattr(change_rates, 'dropna'):
                valid_rates = change_rates.dropna()
            else:
                valid_rates = change_rates
            
            if len(valid_rates) > 0:
                stats_text = f"""
ğŸ“Š {region} ìƒìŠ¹/í•˜ë½ë¥  í†µê³„ ({base_text}):
â€¢ í‰ê·  ë³€í™”ìœ¨: {valid_rates.mean():+.2f}%
â€¢ ìµœëŒ€ ìƒìŠ¹ë¥ : {valid_rates.max():+.2f}%
â€¢ ìµœëŒ€ í•˜ë½ë¥ : {valid_rates.min():+.2f}%
â€¢ ìµœì¢… ë³€í™”ìœ¨: {valid_rates.iloc[-1]:+.2f}%
â€¢ í‘œì¤€í¸ì°¨: {valid_rates.std():.2f}%
â€¢ ë°ì´í„° ê°œìˆ˜: {len(valid_rates)}ê°œ
                """
                print(stats_text)
        else:
            # ë‹¤ì¤‘ ì§€ì—­
            print(f"\nğŸ“Š ì„ íƒëœ {len(all_plot_data)}ê°œ ì§€ì—­ ìµœì¢… ë³€í™”ìœ¨ ({base_text}):")
            for region, data, change_rates in all_plot_data:
                if len(change_rates) > 0:
                    if hasattr(change_rates, 'iloc'):
                        final_rate = change_rates.iloc[-1]
                    else:
                        final_rate = change_rates[-1]
                    print(f"â€¢ {region}: {final_rate:+.2f}%")
            
    def show_statistics(self, price_data, region_name):
        """ì ˆëŒ€ê°€ê²© í†µê³„ ì •ë³´ í‘œì‹œ"""
        stats_text = f"""
ğŸ“Š {region_name} í‰ê· í‰ë‹¨ê°€ í†µê³„ (1í‰ ê¸°ì¤€):
â€¢ í‰ê· : {price_data.mean():.1f}ë§Œì›/í‰
â€¢ ìµœê³ : {price_data.max():.1f}ë§Œì›/í‰
â€¢ ìµœì €: {price_data.min():.1f}ë§Œì›/í‰
â€¢ í‘œì¤€í¸ì°¨: {price_data.std():.1f}ë§Œì›/í‰
â€¢ ë°ì´í„° ê°œìˆ˜: {len(price_data)}ê°œ
        """
        print(stats_text)

    def on_selection_mode_change(self):
        """ì„ íƒ ëª¨ë“œ ë³€ê²½ ì‹œ UI ì—…ë°ì´íŠ¸"""
        mode = self.selection_mode.get()
        
        if mode == "single":
            self.region_combo.configure(state="readonly")
            self.region_listbox.configure(state="normal")
            self.selected_regions_var.set("ìœ„ì˜ ë“œë¡­ë°•ìŠ¤ì—ì„œ ì„ íƒí•˜ì„¸ìš”")
        else:
            self.region_combo.configure(state="readonly")
            self.region_listbox.configure(state="normal")
            self.selected_regions_var.set("ì•„ë˜ ë¦¬ìŠ¤íŠ¸ì—ì„œ Ctrl+í´ë¦­ìœ¼ë¡œ ë‹¤ì¤‘ ì„ íƒí•˜ì„¸ìš”")

    def on_listbox_select(self, event):
        """ë¦¬ìŠ¤íŠ¸ë°•ìŠ¤ ì„ íƒ ì‹œ ì„ íƒëœ ì§€ì—­ í‘œì‹œ ì—…ë°ì´íŠ¸"""
        if self.selection_mode.get() == "multiple":
            selected_indices = self.region_listbox.curselection()
            selected_regions = [self.region_listbox.get(i) for i in selected_indices]
            
            if selected_regions:
                if len(selected_regions) <= 3:
                    self.selected_regions_var.set(", ".join(selected_regions))
                else:
                    self.selected_regions_var.set(f"{', '.join(selected_regions[:3])} ë“± {len(selected_regions)}ê°œ")
            else:
                self.selected_regions_var.set("ì—†ìŒ")

    def save_settings(self):
        """í˜„ì¬ ì„¤ì •ì„ íŒŒì¼ì— ì €ì¥"""
        try:
            settings = {
                'last_file_path': self.file_path_var.get(),
                'last_selected_region': self.region_var.get() if hasattr(self, 'region_var') else '',
                'window_geometry': self.root.geometry(),
                'chart_type': self.chart_type.get() if hasattr(self, 'chart_type') else 'change_rate',
                'base_point': self.base_point.get() if hasattr(self, 'base_point') else 'first',
                'base_year': self.base_year_var.get() if hasattr(self, 'base_year_var') else '2020'
            }
            
            with open(self.config_file, 'w', encoding='utf-8') as f:
                json.dump(settings, f, ensure_ascii=False, indent=2)
                
        except Exception as e:
            print(f"ì„¤ì • ì €ì¥ ì¤‘ ì˜¤ë¥˜: {e}")

    def load_settings(self):
        """ì €ì¥ëœ ì„¤ì •ì„ ë¶ˆëŸ¬ì™€ì„œ ì ìš©"""
        try:
            if os.path.exists(self.config_file):
                with open(self.config_file, 'r', encoding='utf-8') as f:
                    settings = json.load(f)
                
                # íŒŒì¼ ê²½ë¡œ ë³µì›
                last_file_path = settings.get('last_file_path', '')
                if last_file_path and os.path.exists(last_file_path):
                    self.file_path_var.set(last_file_path)
                    
                    # ìë™ìœ¼ë¡œ ë°ì´í„° ë¡œë“œ ì‹œë„
                    try:
                        self.load_data_silent()
                        
                        # ë§ˆì§€ë§‰ ì„ íƒëœ ì§€ì—­ ë³µì›
                        last_region = settings.get('last_selected_region', '')
                        if last_region and hasattr(self, 'region_list') and last_region in self.region_list:
                            self.region_var.set(last_region)
                            
                    except Exception as e:
                        print(f"ìë™ ë°ì´í„° ë¡œë“œ ì‹¤íŒ¨: {e}")
                
                # ì°¨íŠ¸ ì„¤ì • ë³µì›
                chart_type = settings.get('chart_type', 'change_rate')
                if hasattr(self, 'chart_type'):
                    self.chart_type.set(chart_type)
                    
                base_point = settings.get('base_point', 'first')
                if hasattr(self, 'base_point'):
                    self.base_point.set(base_point)
                
                base_year = settings.get('base_year', '2020')
                if hasattr(self, 'base_year_var'):
                    self.base_year_var.set(base_year)
                
                # ì°½ í¬ê¸° ë³µì›
                window_geometry = settings.get('window_geometry', '')
                if window_geometry:
                    try:
                        self.root.geometry(window_geometry)
                    except:
                        pass
                        
        except Exception as e:
            print(f"ì„¤ì • ë¡œë“œ ì¤‘ ì˜¤ë¥˜: {e}")

    def load_data_silent(self):
        """ìë™ ë¡œë“œìš© - ì˜¤ë¥˜ ë©”ì‹œì§€ ì—†ì´ ë°ì´í„° ë¡œë“œ"""
        file_path = self.file_path_var.get()
        if not file_path or not os.path.exists(file_path):
            return False
            
        try:
            # íŒŒì¼ ê²½ë¡œ ì •ê·œí™” ë° ê¶Œí•œ í™•ì¸
            if not os.access(file_path, os.R_OK):
                return False
            
            # ì—‘ì…€ íŒŒì¼ì˜ ì‹œíŠ¸ ëª©ë¡ í™•ì¸
            xl_file = pd.ExcelFile(file_path)
            sheet_names = xl_file.sheet_names
            
            # '47.ã¡ë‹¹ì•„íŒŒíŠ¸í‰ê· ë§¤ë§¤' ì‹œíŠ¸ í™•ì¸
            target_sheet = None
            for sheet in sheet_names:
                if '47' in sheet and 'ã¡' in sheet and 'ì•„íŒŒíŠ¸' in sheet:
                    target_sheet = sheet
                    break
                    
            if target_sheet is None:
                # ì²« ë²ˆì§¸ ì‹œíŠ¸ ì‚¬ìš©
                target_sheet = sheet_names[0] if sheet_names else None
                
            if target_sheet is None:
                return False
            
            # í•´ë‹¹ ì‹œíŠ¸ ì½ê¸°
            self.df = pd.read_excel(file_path, sheet_name=target_sheet)
            
            # ë°ì´í„° ì „ì²˜ë¦¬
            self.preprocess_data()
            
            # ì§€ì—­ ì½¤ë³´ë°•ìŠ¤ ì—…ë°ì´íŠ¸
            self.update_region_combos()
            
            print(f"ìë™ ë¡œë“œ ì™„ë£Œ: {target_sheet} ({len(self.df)}ê°œ í–‰)")
            return True
            
        except Exception as e:
            print(f"ìë™ ë¡œë“œ ì‹¤íŒ¨: {e}")
            return False

    def on_closing(self):
        """í”„ë¡œê·¸ë¨ ì¢…ë£Œ ì‹œ í˜¸ì¶œ"""
        self.save_settings()
        self.root.destroy()

def main():
    root = tk.Tk()
    app = ApartmentPriceAnalyzer(root)
    root.mainloop()

if __name__ == "__main__":
    main()
```
